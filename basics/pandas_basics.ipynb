## 1. 라이브러리 Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries imported!")
```

## 2. 데이터 로딩

```python
# CSV 파일 로딩
df = pd.read_csv('../mock_exams/data/mock_01_delivery.csv')

# 상위 5행 확인
df.head()
```

## 3. 데이터 크기 확인

```python
# shape: (행, 열)
print(f"Shape: {df.shape}")
print(f"행 개수: {df.shape[0]}")
print(f"열 개수: {df.shape[1]}")
```

## 4. 데이터 정보 확인

```python
# info(): 데이터 타입, 결측치 개수 등
df.info()
```

## 5. 기초 통계량

```python
# describe(): 수치형 컬럼의 기초 통계
df.describe()
```

```python
# 범주형 포함 통계
df.describe(include='all')
```

## 6. 결측치 확인

```python
# 컬럼별 결측치 개수
print(df.isna().sum())
```

```python
# 총 결측치 개수
print(f"총 결측치: {df.isna().sum().sum()}")
```

```python
# 결측치 비율 (%)
print(df.isna().mean() * 100)
```

## 7. 결측치 처리

```python
# 복사본 생성 (원본 보존)
df_clean = df.copy()

# 수치형 컬럼: 평균값으로 대체
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    if df_clean[col].isna().sum() > 0:
        df_clean[col].fillna(df_clean[col].mean(), inplace=True)
        print(f"{col}: 평균값으로 대체")
```

```python
# 범주형 컬럼: 최빈값으로 대체
categorical_cols = df_clean.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df_clean[col].isna().sum() > 0:
        df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)
        print(f"{col}: 최빈값으로 대체")
```

```python
# 결측치 처리 확인
print("처리 후 결측치:")
print(df_clean.isna().sum())
```

## 8. 값 빈도 확인

```python
# 범주형 컬럼 값 빈도
df_clean['weather'].value_counts()
```

```python
# 비율로 확인
df_clean['traffic'].value_counts(normalize=True)
```

## 9. 데이터 필터링

```python
# 조건 필터링
df_filtered = df_clean[df_clean['delivery_time'] > 30]
print(f"30분 이상 배달: {len(df_filtered)}건")
```

```python
# 여러 조건
df_filtered = df_clean[(df_clean['weather'] == 'Rain') & (df_clean['traffic'] == 'High')]
print(f"비오고 교통혼잡: {len(df_filtered)}건")
```

## 10. 데이터 정렬

```python
# 오름차순 정렬
df_sorted = df_clean.sort_values('delivery_time')
df_sorted.head()
```

```python
# 내림차순 정렬
df_sorted = df_clean.sort_values('delivery_time', ascending=False)
df_sorted.head()
```

## 11. 그룹별 통계

```python
# 그룹별 평균
df_clean.groupby('weather')['delivery_time'].mean()
```

```python
# 여러 통계
df_clean.groupby('traffic')['delivery_time'].agg(['mean', 'std', 'count'])
```

## 12. 상관관계

```python
# 상관계수 행렬
numeric_df = df_clean.select_dtypes(include=[np.number])
numeric_df.corr()
```

```python
# 특정 컬럼과의 상관관계
numeric_df.corr()['delivery_time'].sort_values(ascending=False)
```

## 13. 시각화

```python
# 히스토그램
plt.figure(figsize=(10, 6))
sns.histplot(df_clean['delivery_time'], kde=True)
plt.title('Delivery Time Distribution')
plt.show()
```

```python
# 상관관계 히트맵
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 박스플롯
plt.figure(figsize=(10, 6))
sns.boxplot(x='weather', y='delivery_time', data=df_clean)
plt.title('Delivery Time by Weather')
plt.show()
```

## 14. 인코딩

```python
# One-Hot Encoding
cat_cols = df_clean.select_dtypes(include=['object']).columns.tolist()
df_encoded = pd.get_dummies(df_clean, columns=cat_cols)

print(f"인코딩 전: {df_clean.shape}")
print(f"인코딩 후: {df_encoded.shape}")
```

```python
df_encoded.head()
```

## 15. 파생변수 생성

```python
# 거리 계산
df_clean['distance'] = np.sqrt(
    (df_clean['delivery_lat'] - df_clean['restaurant_lat'])**2 +
    (df_clean['delivery_lng'] - df_clean['restaurant_lng'])**2
)

df_clean[['restaurant_lat', 'restaurant_lng', 'delivery_lat', 'delivery_lng', 'distance']].head()
```
