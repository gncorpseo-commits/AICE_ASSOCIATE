# 데이터프레임 집중 트레이닝 - 데이터셋 로더

이 노트북은 로컬 경로와 GitHub Raw URL을 자동으로 전환해 CSV를 불러옵니다.

- 로컬에서 실행하면 로컬 경로를 사용합니다.
- Colab에서 실행하면 GitHub Raw URL을 사용합니다.
- 데이터셋 변경 시 `LOCAL_PATH`/`REMOTE_URL`만 바꿔서 사용하세요.

```python
import os
import pandas as pd

# 사용할 데이터셋 경로를 선택하세요
LOCAL_PATH = "../practice/datasets/iris.csv"
REMOTE_URL = "https://raw.githubusercontent.com/gncorpseo-commits/AICE_ASSOCIATE/main/practice/datasets/iris.csv"
DATASET_PATH = LOCAL_PATH if os.path.exists(LOCAL_PATH) else REMOTE_URL

df = pd.read_csv(DATASET_PATH)
df.head()
```
