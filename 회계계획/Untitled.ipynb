```python
import pandas as pd

# 예시 데이터
data = {
    "날짜": ["2026-01-01", "2026-01-03", "2026-01-05", "2026-01-10"],
    "구분": ["수입", "지출", "지출", "수입"],
    "항목": ["월급", "식비", "교통비", "투자수익"],
    "금액": [3000000, 12000, 2500, 50000],
    "결제수단": ["계좌이체", "카드", "현금", "계좌이체"]
}

df = pd.DataFrame(data)
df["날짜"] = pd.to_datetime(df["날짜"])
df
```

```python
import ipywidgets as widgets
from IPython.display import display

항목 = widgets.Text(description="항목")
금액 = widgets.IntText(description="금액")
구분 = widgets.Dropdown(options=["수입", "지출"], description="구분")

display(항목, 금액, 구분)
```
